### Joins

```python
a= pd.DataFrame({'id':[1,2,3],'name':['meena','jhilmil','karan'],'marks':[45,65,89]})
b=pd.DataFrame({'id':[4,5,6,2],'grade':['C','B','A','A']})
```

```python
a
```

```python
b
```

```python
# inner join or intersection of set to get common value
pd.merge(a,b,on='id',how='inner')
```

```python
# left join
pd.merge(a,b,on='id',how='left')
```

```python
# right join
pd.merge(a,b,on='id',how='right')
```

```python
# full join
pd.merge(a,b,on='id',how='outer')
```
